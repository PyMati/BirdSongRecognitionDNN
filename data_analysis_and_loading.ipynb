```python
import pandas as pd
import os
import numpy as np
```

```python
data = {}
for klass in os.listdir("data"):
    if os.path.isdir("data/" + klass):
        data[klass] = os.listdir("data/" + klass)
```

```python
df = pd.read_csv("data/train_mod.csv")
for col in df.columns:
    print(col)
```

```python
data.keys()
```

```python
df = df[df["ebird_code"].isin(os.listdir("data"))]
```

```python
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
print("Species distribution:")
print(df["species"].value_counts())
```

```python
print("\nSong vs Call distribution:")
print(df.groupby("type")["type"].count())
```

```python
print("\nBird seen vs not seen:")
print(df.groupby("bird_seen")["bird_seen"].count())
```

```python
print("\nTop 5 longest recordings:")
print(df.sort_values("duration", ascending=False)["duration"].head())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns


plt.figure(figsize=(12, 6))
sns.countplot(x="species", data=df)
plt.title("Species Distribution")
plt.xticks(rotation=90)
plt.show()
```
